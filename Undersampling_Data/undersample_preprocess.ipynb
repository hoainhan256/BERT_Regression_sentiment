```python
import pandas as pd
from sklearn.utils import resample
```

```python
train_df = pd.read_csv("train.csv")

print("Phân bố nhãn ban đầu:")
train_df["stars"].value_counts()
```

```python
min_count = train_df["stars"].value_counts().min()
min_count
```

```python
balanced_dfs = []
for label in train_df["stars"].unique():
    df_label = train_df[train_df["stars"] == label]
    df_label_down = resample(df_label, 
                              replace=False, 
                              n_samples=min_count, 
                              random_state=42)
    balanced_dfs.append(df_label_down)
train_under = pd.concat(balanced_dfs).sample(frac=1, random_state=42)

```

```python
print("Phân bố nhãn sau khi undersampling:")
train_under["stars"].value_counts()
```

```python
train_under.to_csv("train_undersample.csv", index=False)
print("Đã lưu file")
```

```python
compare_df = pd.DataFrame({
    "Original": train_df["stars"].value_counts().sort_index(),
    "Undersample": train_under["stars"].value_counts().sort_index()
})

compare_df
```
